### The Decoy Effect analysis

The Decoy Effect, first demonstrated in 1982 by Joel Huber and others at Duke University, explains how when a customer is hesitating between two options, presenting them with a third “asymmetrically dominated” option that acts as a decoy will strongly influence which decision they make. An option can be defined as asymmetrically dominated when it is completely dominated by (i. e. definitely inferior to) one option and only partially dominated (i. e. inferior in some aspects) by the other. The asymmetrically dominated option is a decoy serving to increase preference for the dominating option – the one we really want the consumer to choose. Decoys are about asymmetrically quality and price difference, as The Economist does with its subscriptions.

### Example 1: The Economist
One of the best known examples of the decoy effect is an old subscription page of The Economist. They offered 3 different types of subscription:

- Web Subscription - 59  U.S. Dollar 
- Print Subscription – 125  U.S. Dollar
- Web and Print Subscription – 125  U.S. Dollar

The first offer of 59  U.S. Dollars seemed reasonable. The second option (only print) seemed a bit expensive, but still ok. But what about the third option? Both Web and Print for the same price as the print-only subscription?

Dan Ariely, an Israeli American professor of psychology and behavioral economics and author of “Predictably Irrational“, tested this phenomenon with his MIT students where he asked them to choose a subscription. The results were:

- Web Subscription – 59  U.S. Dollar (16 students)
- Print Subscription – 125  U.S. Dollar (0 students)
- Web and Print Subscription – 125  U.S. Dollar (84 students)
Total revenue: 11,444

 
The majority of students selected the third option (dominating) and none of them selected the second option (the decoy). Knowing this, Ariely performed a second test and removed the decoy product. The results were:

- Web Subscription – 59  U.S. Dollar(68 students)
- Web and Print Subscription – 125  U.S. Dollar (32 students)
Total revenue: $8,012

This time, most of the students preferred the first subscription. By adding a decoy product, The Economist improved sales with 30%.

A decoy can also affect your views on political figures.

While Barack Obama was in office, you may not have liked George W. Bush. But now that Donald Trump is in office, you probably like Bush a little more now. In neither case do you like Donald Trump. That’s because Trump is the decoy. Simply adding Donald Trump to the set doesn’t change your views of either Obama or Trump, but it may shift–or even reverse–your opinion of Bush.

Polling data bears this out. Bush’s approval ratings have steadily climbed since Trump took office. Even Democrats now view Bush favorably.

### T-test to indetify if there is significant difference of spendings
As we saw, there is difference of spendings, but is that significant?

- H0: (mu1 - mu2) = 0, there is no difference
- HA: (mu1 - mu2) != 0

```python
a = df['G1_Popcorn($)']
b = df['G2_Popcorn($)']
tStat, pValue = stats.ttest_ind(a, b, equal_var = False) #T-Test
print(f'P-Value:', pValue)
print(f'T-Statistic:', tStat) #print the P-Value and the T-Statistics
```

Our p-value of 0,03 is smaller than 0,05, so we can reject our null hypothesis, there is significant difference of anxiety
